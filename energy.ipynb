```python
# import important libraries 
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
#creating dataframe for years 2015 and above
final_df=df[df['year']>=2015]
final_df.head()
```

```python
final_df.shape
```

```python
# because dataframe contains data for both countries and regions, we will write code to divide them
df_regions=final_df.copy()
```

```python
regions=['Africa','Asia Pacific', 'CIS', 'Central America',
        'Eastern Africa', 'Europe', 'Europe (other)',
        'Middle Africa', 'Middle East',
       'North America', 'OPEC', 'Other Asia & Pacific', 'Other CIS',
       'Other Caribbean', 'Other Middle East', 'Other Northern Africa',
       'Other South America', 'Other Southern Africa',
       'South & Central America', 'World', 'Western Sahara',
       'Western Africa' ]

for i in df_regions['country'].unique():
    if i not in regions:
        df_regions.drop(df_regions[df_regions['country']==i].index, inplace=True)

df_regions.head()
```

```python
#dataframe for countries
df_countries=final_df.copy()
```

```python
for i in df_countries['country'].unique():
    if i in regions:
        df_countries.drop(df_countries[df_countries['country']==i].index, inplace=True)

df_countries.head()
```

```python
#save to csv file for dashboard creation
df_countries.to_csv('df_countries.csv')
```
